```python
import pytesseract
from pytesseract import Output
from PIL import Image, ImageDraw, ImageFont

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
img_loc = r'C:\Users\Habram\Documents\Datasets\SER_custom\AgderEnergiSolutionsGmbH.tif'
img = Image.open(img_loc)

#print(pytesseract.image_to_string(Image.open(img_loc)))

raw_ocr = pytesseract.image_to_data(img, output_type=Output.DICT)
```

```python
words = dict()

for i in range(len(raw_ocr['text'])):
    right = raw_ocr['left'][i] + raw_ocr['width'][i]
    bottom = raw_ocr['top'][i] + raw_ocr['height'][i]
    words[raw_ocr['text'][i]] = [raw_ocr['left'][i], raw_ocr['top'][i], right, bottom]
```

```python
draw = ImageDraw.Draw(img)

for label in words:
    draw.rectangle(words[label])

img
```
