### Visualize block fields

```python
from draw_table_field import LayoutManager
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

layout_manager = LayoutManager()
# Open the blank, empty invoice image
canvas = Image.open(r'blank.tif')
drawer = ImageDraw.Draw(canvas)
font = ImageFont.truetype('arial.ttf', 60)

fig, axs = plt.subplots(1,3)
fig.set_figheight(15)
fig.set_figwidth(30)

for idx in range(3):
    canvas = Image.open(r'blank.tif')
    drawer = ImageDraw.Draw(canvas)
    random_layout = layout_manager.get_layout()
    updated_canvas = layout_manager.visualize_layout(random_layout, drawer, font, canvas)
    axs[idx].imshow(updated_canvas)
    axs[idx].set_title(f'Example {idx}')
```
