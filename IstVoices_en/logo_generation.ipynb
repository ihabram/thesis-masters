100 random shapes, with random color and orientation were generated with this tool: https://github.com/elkorchi/2DGeometricShapesGenerator

100 other images were generated from these images by combining them and getting wider images.

### Show simple shape images

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

def display_images(directory):
    if not os.path.exists(directory):
        print(f"Directory '{directory}' does not exist.")
        return

    # Get a list of all files in the directory
    files = os.listdir(directory)

    # Filter the list to include only image files (you can modify the extensions as needed)
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

    # Randomly select 4 images
    selected_images = random.sample(image_files, 4)

    # Create a figure with a single row and 4 columns
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))

    # Iterate over the selected images and display them
    for i, image_file in enumerate(selected_images):
        # Open the image using PIL
        image_path = os.path.join(directory, image_file)
        image = Image.open(image_path)

        # Display the image on the corresponding subplot
        axes[i].imshow(image)
        axes[i].axis('off')

    # Adjust the spacing between subplots
    plt.subplots_adjust(wspace=0.1)

    # Show the figure
    plt.show()

# Specify the directory containing the images
directory = r"C:\Users\Habram\Documents\Datasets\Company_logos"

# Call the function to display the images
display_images(directory)
```

### Generate concatenated images (combining two simple shapes)

```python
def concatenate_images(directory, output_directory, num_iterations=100):
    if not os.path.exists(directory):
        print(f"Directory '{directory}' does not exist.")
        return

    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Get a list of all files in the directory
    files = os.listdir(directory)

    # Filter the list to include only image files (you can modify the extensions as needed)
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))]

    # Randomly select 2 images and concatenate them num_iterations times
    for i in range(num_iterations):
        # Randomly select 2 images
        selected_images = random.sample(image_files, 2)

        # Open the two selected images using PIL
        images = [Image.open(os.path.join(directory, image_file)) for image_file in selected_images]

        # Get the dimensions of the first image
        width1, height1 = images[0].size

        # Get the dimensions of the second image
        width2, height2 = images[1].size

        # Calculate the width and height of the concatenated image
        concatenated_width = width1 + width2
        concatenated_height = max(height1, height2)

        # Create a new blank image with the concatenated dimensions
        concatenated_image = Image.new('RGB', (concatenated_width, concatenated_height))

        # Paste the first image on the left side of the concatenated image
        concatenated_image.paste(images[0], (0, 0))

        # Paste the second image on the right side of the concatenated image
        concatenated_image.paste(images[1], (width1, 0))

        # Determine the output filename for the concatenated image
        output_filename = f"{i+100}.jpg"
        output_path = os.path.join(output_directory, output_filename)

        # Save the concatenated image to the output path
        concatenated_image.save(output_path)

# Specify the directory containing the images
directory = r"C:\Users\Habram\Documents\Datasets\Company_logos"

# Specify the output directory for the concatenated images
output_directory = r"C:\Users\Habram\Documents\Datasets\Company_logos"

# Call the function to perform concatenation 100 times and save the results
concatenate_images(directory, output_directory, num_iterations=100)
```

### Show concatenated examples

```python
import os
import random
import re
import matplotlib.pyplot as plt
from PIL import Image

def display_images(directory):
    if not os.path.exists(directory):
        print(f"Directory '{directory}' does not exist.")
        return

    # Get a list of all files in the directory
    files = os.listdir(directory)

    # Filter the list to include only image files with names containing a number 100 or greater
    image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')) and re.search(r'\d{3,}', file)]

    # Randomly select 3 images
    selected_images = random.sample(image_files, 3)

    # Create a figure with a single row and 4 columns
    fig, axes = plt.subplots(1, 3, figsize=(20, 5))

    # Iterate over the selected images and display them
    for i, image_file in enumerate(selected_images):
        # Open the image using PIL
        image_path = os.path.join(directory, image_file)
        image = Image.open(image_path)

        # Display the image on the corresponding subplot
        axes[i].imshow(image)
        axes[i].axis('off')

    # Adjust the spacing between subplots
    plt.subplots_adjust(wspace=0.1)

    # Show the figure
    plt.show()

# Specify the directory containing the images
directory = r"C:\Users\Habram\Documents\Datasets\Company_logos"

# Call the function to display the images
display_images(directory)
```
